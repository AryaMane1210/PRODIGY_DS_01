```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
fifa = pd.read_csv('/content/fifa_data.csv')
fig, ax = plt.subplots()
data = fifa.Stamina

N, bins, patches = ax.hist(data, edgecolor='white', linewidth=1)

for i in range(0,3):
    patches[i].set_facecolor('red')
for i in range(3,5):
    patches[i].set_facecolor('yellow')
for i in range(5, len(patches)):
    patches[i].set_facecolor('green')

plt.xlabel("Stamina", fontdict={ 'fontsize': 10})
plt.ylabel("Number of Players", fontdict={ 'fontsize': 10})
plt.xticks(bins)
plt.title("Stamina Distribution", fontdict={ 'fontsize': 20})

plt.show()
```

```python
fig, ax = plt.subplots()
data = fifa.Balance

N, bins, patches = ax.hist(data, edgecolor='white', linewidth=1)

for i in range(0,3):
    patches[i].set_facecolor('green')
for i in range(3,5):
    patches[i].set_facecolor('blue')
for i in range(5, len(patches)):
    patches[i].set_facecolor('orange')

plt.xlabel("Balance", fontdict={ 'fontsize': 10})
plt.ylabel("Number of Players", fontdict={ 'fontsize': 10})
plt.xticks(bins)
plt.title("Balance Distribution", fontdict={ 'fontsize': 20})

plt.show()
```

```python
data= fifa.Volleys
# Plot histogram with density
plt.hist(data, bins=30, density=True, histtype='step')
plt.title('Volley Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.show()
```
